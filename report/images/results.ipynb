# Project Results

```python
import matplotlib.pyplot as plt

# Data
results = {
    "base": (0.62, 0.24),
    "ft": (0.74, 0.61),
    "ft-combined": (0.72, 0.6),
    "ft-expensive": (0.74, 0.58),
    "ft-expensive-combined": (0.74, 0.6),
    "ft-augmented": (0.0, 0.0),
    "ft-augmented-combined": (0.0, 0.0)
}

steps = ["Base", "Standard LoRA", "High Capacity", "Augmentation"]
individual_keys = ["base", "ft", "ft-expensive", "ft-augmented"]
combined_keys = ["base", "ft-combined", "ft-expensive-combined", "ft-augmented-combined"]

# Extracting values
mcq_ind = [results[k][0] for k in individual_keys]
saq_ind = [results[k][1] for k in individual_keys]
mcq_comb = [results[k][0] for k in combined_keys]
saq_comb = [results[k][1] for k in combined_keys]

# Setup
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8), sharex=True, sharey=True)

# Plotting Task-Specific
ax1.plot(steps, mcq_ind, marker='o', label='MCQ Accuracy', color='#1f77b4', linewidth=2)
ax1.plot(steps, saq_ind, marker='s', label='SAQ Accuracy', color='#ff7f0e', linewidth=2)
ax1.set_title('Standard Models (Task-Specific Specialists)', fontweight='bold')
ax1.set_ylabel('Accuracy Score')
ax1.grid(True, linestyle='--', alpha=0.6)
ax1.legend()

# Plotting Combined
ax2.plot(steps, mcq_comb, marker='o', label='MCQ Accuracy', color='#1f77b4', linewidth=2)
ax2.plot(steps, saq_comb, marker='s', label='SAQ Accuracy', color='#ff7f0e', linewidth=2)
ax2.set_title('Combined Models (Multi-Task Architecture)', fontweight='bold')
ax2.set_ylabel('Accuracy Score')
ax2.set_xlabel('Iterative Improvement Steps')
ax2.grid(True, linestyle='--', alpha=0.6)
ax2.legend()

# Annotate points
def annotate_line(ax, data):
    for i, val in enumerate(data):
        ax.annotate(f'{val:.2f}', (i, val), textcoords="offset points", xytext=(0,10), ha='center', fontsize=9)

annotate_line(ax1, mcq_ind)
annotate_line(ax1, saq_ind)
annotate_line(ax2, mcq_comb)
annotate_line(ax2, saq_comb)

plt.tight_layout()
plt.savefig('learning_curve_progression.png')
```
