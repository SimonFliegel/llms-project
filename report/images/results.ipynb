```python
import matplotlib.pyplot as plt

# Data
results = {
    "base": (0.62, 0.24),
    "ft": (0.71, 0.58),
    "ft-combined": (0.72, 0.6),
    "ft-combined-complex": (0.74, 0.6),
    "ft-combined-augmented": (0.79, 0.59)
}

steps = ["Base", "Standard LoRA", "Multi-Task", "High Capacity", "Augmentation"]

mcq = [results[k][0] for k in results.keys()]
saq = [results[k][1] for k in results.keys()]

plt.plot(steps, mcq, marker='o', label='MCQ Accuracy', color='#1f77b4', linewidth=2)
plt.plot(steps, saq, marker='s', label='SAQ Accuracy', color='#ff7f0e', linewidth=2)
plt.title('Accuracy Improvements', fontweight='bold')
plt.ylabel('Accuracy Score')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Plotting Combined
# ax2.plot(steps, mcq_comb, marker='o', label='MCQ Accuracy', color='#1f77b4', linewidth=2)
# ax2.plot(steps, saq_comb, marker='s', label='SAQ Accuracy', color='#ff7f0e', linewidth=2)
# ax2.set_title('Combined Models (Multi-Task Architecture)', fontweight='bold')
# ax2.set_ylabel('Accuracy Score')
# ax2.set_xlabel('Iterative Improvement Steps')
# ax2.grid(True, linestyle='--', alpha=0.6)
# ax2.legend()

# Annotate points
def annotate_line(ax, data):
    for i, val in enumerate(data):
        ax.annotate(f'{val:.2f}', (i, val), textcoords="offset points", xytext=(0,10), ha='center', fontsize=9)

annotate_line(plt.gca(), mcq)
annotate_line(plt.gca(), saq)

plt.tight_layout()
plt.savefig('accuracy_improvements.png')
```
